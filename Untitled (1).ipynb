```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
""" basically this is normalization"""
train_datagen = ImageDataGenerator(rescale=1./255) # rescale the pixel values to [0, 1] 

val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_dir = 'car-damage-detection/data1a/training/'
val_dir = 'car-damage-detection/data1a/validation/'
```

```python
""" Is this Section we are loading the data set so that every image is of same size and batch size is 32 means every batch 
will have 32 images of size 224 x 224, and binary mode used for the two classes"""

train_generator = train_datagen.flow_from_directory(train_dir, target_size=(224, 224), batch_size=32, class_mode='binary')
val_generator = val_datagen.flow_from_directory(val_dir, target_size=(224, 224), batch_size=32, class_mode='binary')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python

"""
now we are adding filters that will be applied to images to extract features and max pooling is reducing the the feature map 
size by factor of 2 , sigmoid is used to return 0,1 is optimal to use with binary category classification.

"""

model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(tf.keras.layers.Dropout(.2))

model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=256, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(units=128, activation='relu'))# 128 neurons
model.add(Dense(units=1, activation='sigmoid',name='bounding_box_and_classes'))
model.summary()
```

```python

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#
history = model.fit(train_generator, epochs=6, validation_data=val_generator)
```
